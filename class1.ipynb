```python
import random

stu1 = list()
for i in range(5):
    stu1.append(random.randint(50,100))
stu1
```

```python
students = []
for _ in range(50):
    stu = list()
    for _ in range(5):
        stu.append(random.randint(50,100))
    students.append(stu)
students
```

```python
import csv
csvfile = open('students.csv',newline='',mode='w',encoding='utf-8')
csva = csv.writer(csvfile)
csva.writerow(['國文','英文','數學','地理','天文'])
csva.writerows(students)
csvfile.close()
```
